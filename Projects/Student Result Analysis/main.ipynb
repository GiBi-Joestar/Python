```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

# Drop Unnamed Column

```python
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
sns.countplot(data=df,x="Gender")
plt.show
```

# From the above chart we have analyed that the number of females in the data is more than the number of males

```python
gp = df.groupby("ParentEduc").agg({"MathScore" : "mean" , "ReadingScore" : "mean" , "WritingScore" : "mean"})
print(gp)
```

```python
sns.heatmap(gp,annot=True)
plt.title("Relationship betweeen parents education and student's Score")
plt.show()
```

# From the above chart we have concluded that parents education have good impact on student's Scores

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore" : "mean" , "ReadingScore" : "mean" , "WritingScore" : "mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship betweeen parents Marital Status and student's score")
```

# From the above chart we have concluded that there is no/negligible impact on student's Scores due to their parent's Marital Status

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f")
plt.title("Distribution of Ethnic Group")
plt.show()
```
